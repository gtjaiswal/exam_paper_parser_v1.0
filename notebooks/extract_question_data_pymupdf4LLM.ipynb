**read document using pymupdf4llm**

```python
import pymupdf4llm

pdf_path = "../data/raw_papers/June 2018 QP - Paper 1 (H) Edexcel Physics GCSE.pdf"
md_text = pymupdf4llm.to_markdown(doc=pdf_path)
print(md_text)
```

**extract chunks**

```python
chunks = pymupdf4llm.to_markdown(doc=pdf_path, page_chunks=True)

import pprint

pprint.pp(chunks)
```

**extract images as files saved seperately**

```python
chunk_images = pymupdf4llm.to_markdown(doc=pdf_path, page_chunks=True, write_images=True, 
                                       image_path="../data/raw_papers/extarcted_images")

pprint.pp(chunk_images)
```

**extract images and embed in the markdown as base64 encoded**

```python
chunk_embeded_images = pymupdf4llm.to_markdown(doc=pdf_path, page_chunks=True, write_images=True, embed_images=True)

pprint.pp(chunk_embeded_images)
```

**extract words**

```python
import pymupdf4llm
import pprint
pdf_path = "../data/raw_papers/June 2018 QP - Paper 1 (H) Edexcel Physics GCSE.pdf"
chunk_words = pymupdf4llm.to_markdown(doc=pdf_path,page_chunks=True, pages=[2],write_images=True, 
                                      image_path="../data/raw_papers/extarcted_images", 
                                      extract_words=True)
pprint.pp(chunk_words)
```
